```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/pcds_data_time/FLNRO-WMB/3.ascii', header=1) # returns DataFrame object

# X = data.iloc[0:-1].values # Purely integer-location based indexing for selection by position.

wind_speed = data['wind_speed'].tolist()
wind_dir = data[' wind_direction'].tolist()
date = data[' time'].tolist()
temp = data[' temperature'].tolist()
rel_humidity = data[' relative_humidity'].tolist()
precip = data[' precipitation'].tolist()

# can you do both of these in one comprehension??
date = [x.strip() for x in date] # strip leading whitespace
date = [x[0:10] for x in date]   # strip time (it's just 00:00:00)

mini = data.head(200)
mini_ws = mini['wind_speed'].tolist()
mini_date = mini[' time'].tolist()
mini_date = [x.strip() for x in mini_date] # strip leading whitespace
mini_date = [x[0:10] for x in mini_date]   # strip time (it's just 00:00:00)


plt.plot(mini_date, mini_ws, 'bo', linewidth=1, label='wind speed')
xlabel = mini_date[::10]
plt.xticks(xlabel, rotation=80)
plt.xlabel('Date')
plt.ylabel('Wind Speed')
plt.legend()
plt.show()
```
